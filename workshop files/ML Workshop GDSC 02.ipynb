To use the Iris dataset in scikit-learn, you will first have to import it.

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset('iris')

# Scatter plot of the sepal length and sepal width
sns.scatterplot(x=iris['sepal_length'], y=iris['sepal_width'], hue=iris['species'], palette='viridis')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Scatter Plot of Sepal Length and Sepal Width')
plt.show()
```

Once the dataset is imported, you can explore it using the describe() method.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25)
```

This will give you a summary of the dataset, including the mean, standard deviation, minimum, and maximum values for each feature.

To split the dataset into training and testing sets, you can use the train_test_split() method.


This will split the dataset into two sets, with 75% of the data in the training set and 25% of the data in the testing set.

To train a model, you can use the fit() method.


```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

To evaluate the model, you can use the score() method.

```python
score = model.score(X_test, y_test)
print(score)
```

This will give you the accuracy of the model on the testing set.
