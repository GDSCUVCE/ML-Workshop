```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Generate some example data
x = np.linspace(0, 10, 20)  # Creating an array of 20 values from 0 to 10
y1 = np.sin(x)               # Computing the sine of each element in x
y2 = np.cos(x)               # Computing the cosine of each element in x
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 8]

```

```python
print(x, y1, y2, categories)
```

```python
# Scatter Plot
plt.figure(figsize=(8, 4))    # Creating a figure with specified size
plt.scatter(x, y1, label='sin(x)', color='blue')  # Scatter plot for sin(x)
plt.scatter(x, y2, label='cos(x)', color='red')   # Scatter plot for cos(x)
plt.title('Scatter Plot')      # Adding a title
plt.xlabel('X-axis')           # Labeling the X-axis
plt.ylabel('Y-axis')           # Labeling the Y-axis
plt.legend()                   # Adding a legend
plt.grid(True)                 # Adding a grid
plt.show()                     # Displaying the plot

```

```python
# Line Plot
plt.figure(figsize=(8, 4))
plt.plot(x, y1, label='sin(x)', color='blue')    # Line plot for sin(x)
plt.plot(x, y2, label='cos(x)', color='red')     # Line plot for cos(x)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Bar Chart
plt.figure(figsize=(8, 4))
plt.bar(categories, values, color=['green', 'orange', 'purple', 'brown'])  # Bar chart
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
